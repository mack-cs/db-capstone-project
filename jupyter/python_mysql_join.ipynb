```python
query = """
SELECT
    FullName,
    ContactNumber,
    TotalCost
FROM
    littlelemon.orders o 
JOIN
    littlelemon.customers c 
ON 
    o.CustomerID = c.CustomerID
WHERE
    TotalCost > 60
"""
```

```python
cursor.execute(query)
```

```python
results = cursor.fetchall()
for item in results:
    print(item)
```
